## Feature Engineering

```python
# Screen Time to Sleep Ratio
genz['Screen_Sleep_Ratio'] = (genz['Screen_Time_Hours'] / (genz['Daily_Sleep_Hours'])).round(2)
```

```python
# make it categorical
genz['Sleep_Risk'] = pd.cut(
    genz['Screen_Sleep_Ratio'],
    bins=[0, 0.5, 1, 2, 100],
    labels=['Low', 'Moderate', 'High', 'Extreme']
)
genz.drop(columns=['Screen_Sleep_Ratio'],inplace=True )
```

```python
genz.head()
```

```python
genz['Sleep_Risk'].value_counts()
```

```python
# Social Media to Real Life Ratio
genz['SocialMedia_RealLife_Ratio'] = (
    genz['Daily_Social_Media_Hours'] / 
    (genz['Study_Work_Hours_per_Day'] + genz['Exercise_Frequency_per_Week'] )
).round(2)
```

```python
genz['SocialMedia_RealLife_Ratio'].max()
```

```python
# make it categorical
genz['Reality_Risk'] = pd.cut(
    genz['SocialMedia_RealLife_Ratio'],
    bins=[0, 0.5, 1, 2, 100],
    labels=['Low Risk', 'Moderate Risk', 'High Risk', 'Extreme Risk']
)
genz['Reality_Risk'].value_counts()
```

```python
genz.drop(columns=['SocialMedia_RealLife_Ratio'],inplace=True)
```

```python
genz['Study_Work_Hours_per_Day'].min()
```

```python
# Productivity Disruption Index
genz['Productivity_Disruption'] = (
    (genz['Daily_Social_Media_Hours'] * genz['Night_Scrolling_Frequency']) / 
    (genz['Study_Work_Hours_per_Day'])
)
```

```python
# make it categorical
genz['Productivity_Risk'] = pd.cut(
    genz['Productivity_Disruption'],
    bins=[-float('inf'), 1, 3, 6, float('inf')],
    labels=['Low', 'Moderate', 'High', 'Extreme']
)
genz['Productivity_Risk'].value_counts()
```

```python
genz.drop(columns=['Productivity_Disruption'], inplace=True)
```

```python
# Emotional Cost per Social Media Hour
genz['Emotional_Cost_Per_Hour'] = (
    (genz['Overthinking_Score'] + genz['Anxiety_Score']) / 
    (2 * (genz['Daily_Social_Media_Hours'] + 0.01))
)
```

```python
# make it categorical
genz['Emotional_Risk'] = pd.cut(
    genz['Emotional_Cost_Per_Hour'],
    bins=[0, 1, 2, 3.5, 100],
    labels=['Low', 'Moderate', 'High', 'Extreme']
)

genz['Emotional_Risk'].value_counts()
```

```python
genz.drop(columns='Emotional_Cost_Per_Hour',inplace= True)
```

```python
genz.head()
```

```python
# Social Media Usage Categories
genz['SM_Usage'] = pd.cut(
    genz['Daily_Social_Media_Hours'],
    bins=[0, 2, 4, 6, 24],
    labels=['Low', 'Moderate', 'High', 'Very High']
)

genz['SM_Usage'].value_counts()
```

```python
genz['Age_Group'] = pd.cut(
    genz['Age'],
    bins=[15, 18, 21, 24, 27],
    labels=['Young_GenZ', 'Core_GenZ', 'Mature_GenZ', 'Older_GenZ']
)

genz['Age_Group'].value_counts()
```

## Distribution Analysis

Many of the columns in the dataset show a **good distribution**, close to normal, which is beneficial for correlation and regression analysis.

However, a few variables show skewness that should be considered:

| Column Name              | Skew Type    | Interpretation / Notes                                                                                                   |
| ------------------------ | ------------ | ------------------------------------------------------------------------------------------------------------------------ |
| **Online_Gaming_Hours**  | Right-skewed | Most participants play few hours, but a small number play extensively. Could affect mean and correlation.                |
| **Caffeine_Intake_Cups** | Right-skewed | Majority consume few cups, while some participants consume high amounts. Consider log transformation if using in models. |
| **Sleep_Quality_Score**  | Left-skewed  | Most participants report high sleep quality, few report low. Could affect statistical assumptions for parametric tests.  |

# Insights from Correlation Matrix Analysis

## **Strong Positive Correlations (Strong Relationships)**

### 1. **Screen Time & Daily_Social_Media_Hours (r = 0.83)**
- **Insight**: Extremely strong correlation between daily social media hours and total screen time
- **Meaning**: Social media is the primary driver of screen time for Gen Z
- **Action**: Interventions targeting screen time should focus on social media specifically

### 2. **Anxiety & Emotional Fatigue (r = 0.50)**
- **Insight**: Moderate-strong correlation between anxiety and emotional exhaustion
- **Meaning**: Anxious individuals are at higher risk of burnout
- **Action**: Early anxiety intervention could prevent emotional fatigue

##  **Moderate Positive Correlations (Notable Patterns)**

### 1. **Night Scrolling & Overthinking (r = 0.42)**
- **Insight**: Night-time social media use correlates with rumination
- **Meaning**: Late-night scrolling may trigger or worsen overthinking
- **Action**: Digital curfews could reduce nighttime rumination

### 2. **Social Media Hours & Overthinking (r = 0.26)**
- **Insight**: More social media use = more overthinking
- **Meaning**: Social media may fuel comparison and rumination
- **Action**: Mindful social media consumption programs needed

##  **Key Negative Correlations**

### 1. **Sleep Quality & Anxiety (r = -0.37)**

- **Meaning**: Sleep disruption and anxiety create a vicious cycle
- **Action**: Sleep hygiene programs could reduce anxiety
